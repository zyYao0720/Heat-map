```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel("图7---IMF之间相关性.xlsx",index_col=0)
df
```

```python
plt.rc('font',family='Times New Roman')
corr_mat = df
f, ax = plt.subplots(figsize=(12, 10))
mask = np.zeros_like(corr_mat)
for i in range(1,len(mask)):
  for j in range(0,i):
    mask[j][i] = True
corr_mat=np.around(corr_mat,decimals=2)
sns.heatmap(corr_mat, annot=True,cmap='GnBu',mask=mask,linewidths=.05,square=False,annot_kws={'size':8})
plt.subplots_adjust(left=.1, right=0.95, bottom=0.22, top=0.95)
plt.savefig("图7.png",dpi=1200,bbox_inches='tight')
```

```python
# 通过窗口的形式显示图片, 很好地体现子图与整个画布的位置关系
%matplotlib inline
# 画布被规划为3×2 的矩阵区域 , 之后在索引为6 的区域中绘制子图

xticklabels111=['IMF1','IMF2','IMF3','IMF4','IMF5','IMF6','IMF7','IMF8','IMF9','IMF10','IMF11']
plt.rc('font',family='Times New Roman')
df1=pd.read_excel("图13(a)---预测结果与真实水位imfs的相关性.xlsx",index_col=0)
df2=pd.read_excel("图13(b)---预测结果与外部变量的相关性.xlsx",index_col=0)

labels1=['EF1','EF2','EF3','EF4','EF5','EF6','EF7','EF8','EF9','EF10','EF11','EF12','EF13','EF14','EF15','EF16','EF17','EF18']
labels2=['EF19','EF20','EF21','EF22','EF23','EF24','EF25','EF26','EF27','EF28','EF29','EF30','EF31','EF32','EF33','EF34','EF35']


ax_one = plt.subplot(131) 
sns.heatmap(df1, annot=True,cmap='GnBu',linewidths=.05,square=False,annot_kws={'size':8},yticklabels=xticklabels111,xticklabels=['(a) '])
plt.subplots_adjust(left=.1, right=0.95, bottom=0.22, top=0.95)


ax_two = plt.subplot(132) 
sns.heatmap(df2.iloc[:18,:], annot=True,cmap='GnBu',linewidths=.05,square=False,annot_kws={'size':8},yticklabels=labels1,xticklabels=['(b) -1'])
plt.subplots_adjust(left=.1, right=0.95, bottom=0.22, top=0.95)

ax_three= plt.subplot(133) 
sns.heatmap(df2.iloc[18:,:], annot=True,cmap='GnBu',linewidths=.05,square=False,annot_kws={'size':8},yticklabels=labels2,xticklabels=['(b) -2'])
plt.subplots_adjust(left=.1, right=0.95, bottom=0.22, top=0.95)
plt.tight_layout()
#plt.show()
plt.savefig("图13_test.png",dpi=1200,bbox_inches='tight')
```
